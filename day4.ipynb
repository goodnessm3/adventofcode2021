```python
with open("input4.txt", "r") as f:
    data = f.read()
d = data.split("\n\n")
seq = d[0].split(",")

def make_grid(astr):
    
    a = []
    b = [[0 for _ in range(5)] for _ in range(5)] # rotated 90 degrees
    rows = astr.split("\n")
    for x in rows:
        p = x.split(" ")
        while "" in p:
            p.remove("")
        a.append(p)
    for x in range(len(a)):
        for y in range(len(a)):
            b[x][y] = a[y][x]
    
    return(a, b)

grids = [make_grid(x) for x in d[1:]]  # need to trim final newline from input file
```

```python
from copy import deepcopy

def time_to_win(grid):
    
    """Returns how soon the grid wins, and the final number called before the win."""
    
    cnt = 0
    i = iter(seq)
    cp = grid
    g1, g2 = cp
    
    while not ([0, 0, 0, 0, 0] in g1 or [0, 0, 0, 0, 0] in g2):
        try:
            m = next(i)
        except StopIteration:
            return 99999
        for g in cp:
            for x in range(5):
                for y in range(5):
                    if g[x][y] == m:
                        g[x][y] = 0
            cnt +=1
    
    score = 0
    for x in range(5):
        for y in range(5):
            score += int(g1[x][y])
    score *= int(m)
        
    return cnt, score     
```

```python
st = []
for x in grids:
    bz = deepcopy(x)
    st.append((time_to_win(x), x, bz))
st.sort(key=lambda x: x[0])
```
